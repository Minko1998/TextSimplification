```python
import pandas as pd

data = pd.read_csv('simpletext_task2_decorated_run.csv')
terms = list(data.term)
```

```python
unigram = []
multigrams = []
for term in terms:
    if len(term.split()) > 1:
        multigrams.append(term)
    else:
        unigram.append(term)
unigram = list(set(unigram))
multigrams = list(set(multigrams))
```

```python
print(len(unigram))
print(len(multigrams))
```

```python
unigram.sort(key=len)
multigrams.sort(key=len)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
lengths = [len(uni) for uni in unigram]
plt.hist(lengths, bins = len(set(lengths)), rwidth = .9)
plt.yticks(range(0,21, 4))
ax.set_xlabel('Number of Letters')
ax.set_ylabel('Frequency')
plt.title("Distribution of amount of letters in answers")
plt.savefig("images/DistributionLettersUnigrams.png")
```
